```python
#3) Depict different Market types and their count on a bar chart???
df1=(df['Market'].value_counts())
print(df1)
df3=df1.index.tolist()
print(df3)
df4=df1.values.tolist()
print(df4)
print('------------------------')
plt.figure(figsize=(8,4))
plt.bar(df3,df4,label="df3 vs df4",color='pink',width=.6)
plt.legend()
plt.title('Market Type')
plt.xlabel('df3')
plt.ylabel('df4')
# plt.xlim(1,5)
plt.ylim(1,1400)
plt.show()
```

```python
#15-Depict the quartile distribution of COGS where Sales is greater than 400.....
psr11=df[['Sales','COGS']]
#print(psr11) 
psr2=psr11[psr11['Sales']>400]
print(psr2)
d=psr2['COGS']
print(d)
# from collections import Counter
# print(len(d))
# c = Counter(d)
# print(c)
print('###################')
# print(len(d))
# print(d)
Q0 = np.percentile(d, 0)    
Q1 = np.percentile(d, 25) 
Q2 = np.percentile(d, 50)
Q3 = np.percentile(d, 75) 
Q4 = np.percentile(d, 100) 
  
IQR = Q3 - Q1 
print('Q0',Q0)
print('Q1',Q1) # 1st quartile
print('Q2',Q2) # meadin
print('Q3',Q3)
print('Q4',Q4)
print('IQR',IQR)
min = Q1 - 1.5 * IQR
max = Q3 + 1.5 * IQR
print(min)
print(max)
plt.boxplot(d)
plt.show()
# plt.boxplot(psr2['COGS'])
# plt.show()
```
